```python
import pandas
import numpy
import matplotlib
import matplotlib.pyplot as plt

data = pandas.read_csv('anime.csv')
```

```python
print(data.head(10))
```

```python
data.columns = map(lambda var: var.lower().replace(' ', '_'), data.columns)
data['voters'] = data['voters'].str.replace(',','')
```

```python
data = data.replace('?', numpy.nan).replace('N/A', numpy.nan).replace('-', numpy.nan)
print(data)
data = data.astype({'title' : 'string', 'production' : 'string', 'episodes' : 'float32', 'source' : 'string', 'genre' : 'string', 'airdate' : 'string', 'rating' : 'float32', 'voters' : 'float32', 'theme' : 'string'})
```

```python
print(data.describe([.25, .50, .75, .90]))
```

```python
print(data['production'].value_counts(), '\n')
print(data['source'].value_counts(), '\n')
print(data['genre'].value_counts(), '\n')
print(data['theme'].value_counts(), '\n')
```

```python
print(data.info())
```

```python
data['production'].value_counts().sort_values().plot.bar()

print(data['production'].value_counts().keys()[0], '-', data['production'].value_counts()[0], '\n')
```

```python
data['episodes'].value_counts().plot.bar()

print(data['episodes'].value_counts().keys()[0], '-', data['episodes'].value_counts().values[0], '\n')
```

```python
data['source'].value_counts().head(3).plot.bar()

print(data['source'].value_counts().keys()[0], '-', data['source'].value_counts().values[0], '\n')
```

```python
data['theme'].value_counts().plot.bar()

print(data['theme'].value_counts().keys()[0], '-', data['theme'].value_counts().values[0], '\n')
```

```python
aboba = pandas.Series(map(lambda var : numpy.nan if var == '????' else var, map(lambda var : (var.split(',')[1].strip()),data['airdate'].dropna()))).value_counts()

aboba.plot.bar()

print(aboba.keys()[0], '-', aboba.values[0], '\n')
```

```python
aboba = data[['production', 'rating']].dropna().groupby('production').mean().sort_values('rating')

aboba.plot.bar()

print(aboba.tail(3).sort_values('rating', ascending = False))
```

```python
data['rating'].plot.hist(bins = range(11))
```

```python
aboba1 = data.assign(genre=data['genre'].str.split(',')).explode('genre')[['genre', 'rating']].groupby('genre').mean()
aboba2 = data.assign(theme=data['theme'].str.split(',')).explode('theme')[['theme', 'rating']].groupby('theme').mean()
print(aboba1)
print(aboba2)

aboba1.sort_values('rating').plot.bar()
aboba2.sort_values('rating').plot.bar()
```
